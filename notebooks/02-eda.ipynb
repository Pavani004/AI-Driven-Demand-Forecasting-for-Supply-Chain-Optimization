```python
import pandas as pd
import numpy as np

# Create sample raw data
data = {
    'date': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'sales': np.random.randint(100, 1000, size=100),
    'product_id': np.random.choice(['A', 'B', 'C'], size=100)
}

# Convert to DataFrame
sales_data = pd.DataFrame(data)

# Save to CSV in the 'data' folder
sales_data.to_csv('data/sales_data.csv', index=False)

print("Sample sales_data.csv file created successfully in the 'data' folder.")
```

```python
import pandas as pd

# Load the raw data
sales_data = pd.read_csv('data/sales_data.csv')

# Clean the data (e.g., handle missing values, remove duplicates)
cleaned_data = sales_data.dropna()  # Drop rows with missing values
cleaned_data = cleaned_data.drop_duplicates()  # Remove duplicate rows

# Preprocess the data (e.g., feature engineering)
cleaned_data['date'] = pd.to_datetime(cleaned_data['date'])  # Convert 'date' to datetime
cleaned_data['year'] = cleaned_data['date'].dt.year  # Extract year
cleaned_data['month'] = cleaned_data['date'].dt.month  # Extract month
cleaned_data['day'] = cleaned_data['date'].dt.day  # Extract day

# Save the cleaned data
cleaned_data.to_csv('data/cleaned_sales_data.csv', index=False)

print("Cleaned data saved to 'data/cleaned_sales_data.csv'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the cleaned data
cleaned_data = pd.read_csv('data/cleaned_sales_data.csv')

# Plot sales trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='sales', data=cleaned_data)
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```
